```python
import os
import cv2
import numpy as np
from keras import Sequential
from keras.layers import Flatten, Dense
from sklearn.model_selection import train_test_split

IMG_WIDTH = 128
IMG_HEIGHT = 128
TEST_RATIO = 0.2

def load_images(directory, label):
    images = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.jpg'):
                img_path = os.path.join(root, file)
                img = cv2.imread(img_path)
                img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
                images.append(img.flatten())
    labels = [label] * len(images)
    return images, labels

# Load the dataset
X_normal, y_normal = load_images('databasesepia/normal', 0)
X_sepia, y_sepia = load_images('databasesepia/sepia', 1)

# Combine the data and labels
X = np.array(X_normal + X_sepia)
y = np.array(y_normal + y_sepia)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_RATIO, random_state=42)

# Reshape the data
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Build the neural network model
model = Sequential([
    Dense(64, activation='relu', input_shape=(IMG_WIDTH * IMG_HEIGHT * 3,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print('Test accuracy:', accuracy)

model.summary()

```
